```python
# Q4
# Create a 4x4 NumPy array filled with random integers between 1 and 100. 
# Then, reshape this array into two separate 2D arrays, where one represents the rows and the other represents the columns. 
# Write a function, preferably using a lambda function, to calculate the sum of each row and each column separately, and return the results as two separate NumPy arrays

arr = np.random.randint(1,101,size=(4,4))
rows, cols = arr.shape
sum_func = lambda i : np.array([i.sum(axis=0), i.sum(axis=1)])
row_sum = sum_func(arr.reshape(rows,-1))
col_sum = sum_func(arr.reshape(-1,cols))
print(row_sum)
print(col_sum)
```
